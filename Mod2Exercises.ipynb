### Exercise 3:
Using the lists in the cell below, write code that will zip up the lists and make them into one list, then turn it into a dataframe. Next, export the dataframe as a csv file. Then try exporting the dataframe as an Excel file.

```python
names = ['Nike','Adidas','New Balance','Puma','Reebok']
grades = [176,59,47,38,99]
GradeList = list(zip(names, grades))
```

```python
# Export dataframme to csv file
df = pd.DataFrame(data = GradeList, columns=['Names','Grades'])
df.to_csv('myGradeList.csv', index=False, header=False)
```

```python
# Export dataframe to Excel file
writer = pd.ExcelWriter('myGrades.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1')
writer.save()
```

### Exercise 6:

Using the travel_df dataframe and pivot_table function, get the average total time by day of the week and direction traveled (Home/GSK). 

```python
# Show the sata types of each column
travel_df.dtypes
```

```python
# The average total time by day of week
travel_df['TotalTime'].groupby(travel_df['DayOfWeek']).mean()
```

```python
# The average total time by direction traveled
travel_df.groupby(['GoingTo', 'TotalTime']).mean()
```

```python
# Now use pivot table for average total time by day of week and going to
pd.pivot_table(travel_df, values=['TotalTime'], index=['DayOfWeek', 'GoingTo']) # uses mean by default
```
